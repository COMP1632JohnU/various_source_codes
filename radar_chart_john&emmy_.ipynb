# ***How to create a Radar Chart in Python***
Radar Chart is a visualization technique used to compare multiple varaibles
This is a tutorial on how to create a radar chart. Radar Chart is also called as Spider Chart or Web Chart is a praphical method used
for comparing multiple quantitative variables. It is a two dimensional polar visualization.

# ***Import Libraries***

```python
# using Matplotlib library for visualization and Numpy library for several mathematical functions.
# note that i have set the style of matplotlib as ggplot, set the favourite style or use the default style.
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

# ***Data for Visualization***
Using the marks obtained by 2 students Alice and Bob in 5 subjects using Radar Chart. The subjects and marks are provided as list.

```python
subjects =['PHYSICS','CHEMISTRY','BIOLOGY','MATHEMATICS','ECONOMICS']
john = [70,50,78,80,87]
emmy = [81,38,75,37,46]
```

# ***Obtain Angles***

In the example, the choice of radar chart to visualize 5 variables. Hence the chart would have 5 polar axes.

Total angle within the circle is 360 degre (2 pi radians). the angle between axis will be 72 degrees (360/5).

Matplotlib uses the angles in radians for preparing polar plots. we can obtain the angles using line space function of numpy as below.

Note enpoint parameter has to be set to false.

```python
angles = np.linspace(0,2*np.pi,len(subjects), endpoint=False)
print(angles)
```

# ***Completing full circle***

Now that the angkes for each varaiable is known, this could generate only four-fifth of a circle, the next step is adding the first angle at the end of the array to completely describe the circle.

```python
angles = np.concatenate((angles,[angles[0]]))
print(angles)
```

Similarly the first value has to be appended for all other arrays/lists used for visualisation to maintain length consistency.

```python
subjects.append(subjects[0])
john.append(john[0])
emmy.append(emmy[0])
```

# ***Plotting Radar Chart***
A basic radar chart can be plotted by setting polar = True in add_subplot function, and providing angles and values as argument to plot function.

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(polar = True)
ax.plot(angles, john)
plt.show()
```

# ***Customizing the Radar Chart***

We can customize the radar chart by adding more information.

# ***Line and Marker***

Line and marker styles can be changed by providing arguments within the plot function.
In the below code, line and marker style are set by using positional argument similar to a matplotlib line chart.
Line color is set by using color argument of plot function.

# ***Fill the Plot Area***

Plot are can be filled using the fill function providing the angle and values as arguments.
Color and opacity can be set by using color and alpha parameters. It is a good practice to use contrasting colors and lower value of alpha when comparing for multiple entities.

# ***Adding Polar Grids and Labels***

We can observe in the above figure that the polar axes are spaced at 45 degree from each other and the angles are provided as labels.
The spacing of angles and the labels can be customized by using set_thetagrids function.

# ***Add Legend***

Legend can be added by using plt.legend function and setting label within the plot function.
Add or Remove Grid
Grid function can be used to show/hide the circular grid.

```python
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(polar=True)
#basic plot
ax.plot(angles,john, 'o--', color='g', label='John')
#fill plot
ax.fill(angles, john, alpha=0.25, color='g')
#Add labels
ax.set_thetagrids(angles * 180/np.pi, subjects)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(polar=True)
#basic plot
ax.plot(angles,emmy, 'o--', color='b', label='Emmy')
#fill plot
ax.fill(angles, emmy, alpha=0.25, color='b')
#Add labels
ax.set_thetagrids(angles * 180/np.pi, subjects)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

# ***John vs Emmy***

Finally we have reached the much anticipated section of the tutorial, How John fares against Emmy?
This question can be answered by combining radar charts for John and Emmy in the same chart.

```python
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(111, polar=True)
#John Plot
ax.plot(angles,john, 'o-', color='g', linewidth=1, label='John')
ax.fill(angles, john, alpha=0.25, color='g')
#Emmy Plot
ax.plot(angles,emmy, 'o-', color='orange', linewidth=1, label='Emmy')
ax.fill(angles, emmy, alpha=0.25, color='orange')
ax.set_thetagrids(angles * 180/np.pi, subjects)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```

# ***Finally***

Just by glancing trhough the radar chart, we can observe that John performed much better than Emmy in Economics.

The performance of them were comparable in Biology and Physics.

Johns performance in Mathematics and Chemistry was better than that of Emmy.

This is the advantage of Radar chart, it is very easy to comprehend when comparing several entities on multiple variables.

A radar chart is ideal for comparing 24 entities over 310 quantitative variables. Larger number of entities or variables can render the chart cluttered and difficult to comprehend.
