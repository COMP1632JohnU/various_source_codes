# ***Further Data Pre-Processing***



```python
# axis 0 referes to the row, 1 = column
j5yrs.drop(['description', 
        'policy_reference', 'inception_date',
       'expiry_date',  'territory', 'obligor_trade_sector',
       ], axis=1, inplace=True)
j5yrs.head()
```

# *Encoding the categorical features*

```python
# using one hot encoding fucntion get dummy to convert description contact into numeric data of its own.
pd.concat([j5yrs, pd.get_dummies(j5yrs.description_contact)], axis=1)
```

```python
j5yrs = pd.concat([j5yrs, pd.get_dummies(j5yrs.description_contact)], axis=1)
j5yrs.head()
```

```python
# encoding sex
j5yrs['female']=pd.get_dummies(j5yrs.sex)['Female']
j5yrs['male']=pd.get_dummies(j5yrs.sex)['Male']
j5yrs.head()
```

```python
# dropping the column sex
j5yrs.drop(['sex',], axis=1, inplace=True)
j5yrs.head()
```

```python
# changing the caps from capital letters to small letters
j5yrs['claim'] = j5yrs.claim.str.lower()
j5yrs['bind/ntu'] = j5yrs['bind/ntu'].str.lower()

j5yrs.head()
```

```python
j5yrs.drop(['description_contact',], axis=1, inplace=True)
j5yrs.head()
```

```python
# encoding the column claim & bind/ntu, as well as dropping the claim and bind/ntu after encoding using the panda series get dummies.
pd.concat([j5yrs, pd.get_dummies(j5yrs.claim), pd.get_dummies(j5yrs['bind/ntu'])], axis=1).drop(['claim', 'bind/ntu'], axis=1)
# j5yrs.head()
```

```python
# assigning the encoded columns back to the dataframe
j5yrs = pd.concat([j5yrs, pd.get_dummies(j5yrs.claim), pd.get_dummies(j5yrs['bind/ntu'])], axis=1).drop(['claim', 'bind/ntu'], axis=1)
```

# ***Splitting the Independent and Dependent Variables***

The next step is splitting the Features and Target Variables, which would be the Gross Estmated Written premium, if the instruction to python is to remove a column, mention the axis as 1, and the instruction to python is remove a row, mention the axis as 0.


```python
# to select multiple columns in a dataframe, the list function is best practice where the syntax expected is double square brackets. panda datframe 2 brackets and panda series 1 bracket.
# https://pandas.pydata.org/pandas-docs/stable/getting_started/intro_tutorials/03_subset_data.html
x = j5yrs[['age', 'lloyds_claim_code', 'grosswp', 'a_colclough','d_hanly', 'm_esdaile', 'm_woollam', 'n_de_clermont', 's_james', 'female','male','no','yes','bind', 'ntu']]
y = j5yrs[['grossepi']]
```

# ***Model Trianing Approach = Linear Regression***

```python
# Prediction on training data, calling up and instance of linear regression and assigning it to app.
# https://www.analyticsvidhya.com/blog/2021/05/multiple-linear-regression-using-python-and-scikit-learn/#h2_3
app = LinearRegression()
```

```python
# fitting the data into x_train, y_train
app.fit(x_train,y_train)
```

```python
# finally, to execute this then, the model will be ready, now the  x_test data and use this data for the prediction of forcast. 
forcast = app.predict(x_test)
forcast
```

# ***Model metrics output & Accuracy***

```python
# predicting the accuracy score
# have another model to compare and MLP or neural take a typical neural network found online, and the purpose is for comparison if not code developed.
# 
# https://www.analyticsvidhya.com/blog/2021/05/multiple-linear-regression-using-python-and-scikit-learn/#h2_6
score=r2_score(y_test,forcast)
print('r2 score is ',score)
print('mean_sqrd_error is==',mean_squared_error(y_test,forcast))
print('root_mean_squared error is==',np.sqrt(mean_squared_error(y_test,forcast)))
print('mean_absolute_error is==',mean_absolute_error(y_test,forcast))
```
